```python
# DataFrame을 합치는 방법
# concat
sal_2016 = { '이름' : ['홍', '일', '전'], 
             '급여' : [200, 150, 250]}
sal_2017 = { '이름' : ['일', '강', '전', '홍'],
             '급여' : [180, 210, 270, 220]}
df_sal_2016 = pd.DataFrame(sal_2016, columns = ['이름', '급여'])
df_sal_2017 = pd.DataFrame(sal_2017, columns = ['이름', '급여'])
```

```python
df_sal_2016
```

```python
df_sal_2017
```

```python
sal_concat = pd.concat([df_sal_2016, df_sal_2017])
sal_concat
```

```python
sal_concat_2 = pd.concat([df_sal_2016, df_sal_2017], axis = 1)
sal_concat_2
```

```python
sal_concat_3 = pd.concat([df_sal_2016, df_sal_2017], axis = 1, join = 'inner')
sal_concat_3
```

```python
# DataFrame을 합치는 방법
# merge
sal_merge_4 = pd.merge(df_sal_2016, df_sal_2017, on = '이름')
sal_merge_4
```

```python
sal_merge_5 = pd.merge(df_sal_2016, df_sal_2017, on = '이름', how = 'right')
sal_merge_5
```

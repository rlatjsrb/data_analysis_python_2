```python
import matplotlib.pyplot as plt
# Figure : 그림을 그릴 수 있는 영역, 비어있는 종이
# Subplot : 실제 그림이 그려지는 영역, 종이를 가상으로 나눈 등분
# matplotlib를 그리는 순서
# 1. Figure 설정
# 2. Subplot 작업
```

```python
fig = plt.figure()
ax = fig.add_subplot(111) # 행번호, 열번호, 그림번호
data =[0,1]
ax.plot(data)
plt.show()
```

```python
fig = plt.figure()
x1 = fig.add_subplot(221) # 2행 2열 1번째 그래프
x2 = fig.add_subplot(222) # 2행 2열 2번째 그래프
x3 = fig.add_subplot(223) # 2행 2열 3번째 그래프
plt.show()
```

```python
plt.style.use('ggplot')
fig = plt.figure()
x = fig.add_subplot(111)
data = [0,1]
x.plot(data)
plt.show()
```

```python
#꺾은선 그래프
fig = plt.figure()
ax = fig.add_subplot(111)
x = [0,2,4]
y = [0,4,2]

ax.plot(x, y)
plt.show()
```

```python
# 산포도 그래프(회귀분석 등 특정 패턴을 찾을 때 유용한 그래프)
import numpy as np
plt.style.use('ggplot')
```

```python
# 랜덤
np.random.seed(2)
x = np.arange(1, 101)
y = 2 * x * np.random.rand(100)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(x, y)
plt.show()
```

```python
# 막대그래프
plt.style.use('ggplot')

fig = plt.figure()
ax = fig.add_subplot(111)
x = [3,4]
y = [1,2]

ax.bar(x, y)
plt.show()
```

```python
# 그래프 레이블 생성하기
plt.style.use('ggplot')
fig = plt.figure()
ax = fig.add_subplot(111)
labels = ['english', 'math']
x = [3, 4]
y = [1, 2]
ax.bar(x, y, tick_label = labels)
plt.show()
```

```python
# 가로로 누운 그래프
plt.style.use('ggplot')
fig = plt.figure()
ax = fig.add_subplot(111)

labels = ['english','math']
x = [3,4]
y = [1,2]
ax.barh(x, y, tick_label = labels)
plt.show()
```

```python
# 여러 데이터를 묶어 그리기
x = [1,2]
y1, y2, y3 = [1,2],[2,5],[3,7]

fig = plt.figure()
ax = fig.add_subplot(111)

w = 0.2
ax.bar(x, y1, width = w, label = 'english')
ax.bar(np.array(x) + w, y2, width  = w, label = 'math')
ax.bar(np.array(x) + w * 2, y3, width = w, label = 'korean')

ax.legend()
plt.show()
```

```python
# 히스토그램
plt.style.use('ggplot')

data1 = 100
data2 = 10
np.random.seed(0)
x = np.random.normal(data1, data2, 1000)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.hist(x)
plt.show()
```

```python
# Boxplot
plt.style.use('ggplot')

x = [1,2,3,3,11,20]
fig = plt.figure()
ax = fig.add_subplot(111)
ax.boxplot(x)
plt.show()
```

```python
# Pie 차트
plt.style.use('ggplot')
labels = ['english','math','korean']
sizes = [45, 30, 25]
```

```python
fig = plt.figure(figsize = (3, 3))
ax = fig.add_subplot(111)
ax.pie(sizes, labels = labels)
plt.show()
```
